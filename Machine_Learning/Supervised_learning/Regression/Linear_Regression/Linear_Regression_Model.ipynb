```python
# spliting Independent & Dependent variables using numpy

X=df['cgpa'].values.reshape(-1,1)
Y=df['package '].values.reshape(-1,1)
```

**Train Test Split**


```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.25,random_state=38)
```

```python
print('xtrain',xtrain.shape)
print('xtest',xtest.shape)
print('ytrain',ytrain.shape)
print('ytest',ytest.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
model=lr.fit(xtrain,ytrain)
```

```python
ypred=model.predict(xtest)
```

```python
ytest
```

```python
ypred
```

```python
plt.scatter(X,Y,color='r')
plt.plot(xtest,ypred)
plt.xlabel("cgpa")
plt.ylabel("packages")
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
from sklearn.metrics import mean_squared_error
```

```python
from math import sqrt as sqrt
```

**Mean Absolute Error MAE**

```python
print("MAE",mean_absolute_error(ytest,ypred))
```

**Mean Squared Error MSE**

```python
print("MSE",mean_squared_error(ytest,ypred))
```

**Root Mean Squared Error MSE**

```python
print("RMSE",sqrt(mean_squared_error(ytest,ypred)))
```

```python
# To predict a new value that model has not seen / which was not the part of the training / test set

model.predict([[8.1]])
```
